### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()

# Verificando a estrutura dos dados de todas as lojas
print("Loja 1:")
display(loja.info())

print("\nLoja 2:")
display(loja2.info())

print("\nLoja 3:")
display(loja3.info())

print("\nLoja 4:")
display(loja4.info())

loja.head()
```

## Análise das colunas de cada loja

As quatro lojas possuem o mesmo número de colunas e os mesmos tipos de dados. A única exceção é a Loja 4, que contém uma linha a menos (2358 registros). Nenhum valor nulo foi encontrado, o que permite seguir com a análise normalmente.


#1. Análise do faturamento


Agora que verificamos que as quatro lojas possuem a mesma estrutura de dados, vamos unificar todos os dados em um único DataFrame. Para isso, adicionaremos uma nova coluna chamada `Loja` em cada DataFrame para identificar a origem de cada linha. Essa junção facilitará análises comparativas entre as lojas.**negrito**

```python
# Adiciona uma coluna identificando a loja
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Junta todos os dados
dados = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Faturamento por loja
faturamento = dados.groupby('Loja')['Preço'].sum().sort_values()

# Exibe o faturamento de todas as lojas
print("Faturamento por loja:")
for loja, valor in faturamento.items():
    print(f"{loja}: R${valor:,.2f}")

# Identifica as lojas com menor e maior faturamento
loja_menor = faturamento.index[0]
loja_maior = faturamento.index[-1]

faturamento_menor = faturamento.iloc[0]
faturamento_maior = faturamento.iloc[-1]

# Calcula a diferença percentual
diferenca_percentual = ((faturamento_maior - faturamento_menor) / faturamento_menor) * 100

# Exibe os resultados
print(f"\nLoja com menor faturamento: {loja_menor} - R${faturamento_menor:,.2f}")
print(f"Loja com maior faturamento: {loja_maior} - R${faturamento_maior:,.2f}")
print(f"Diferença percentual do menor para o maior: {diferenca_percentual:.2f}%")

```

# 2. Vendas por Categoria


```python
# # Categoria mais vendida por loja
vendas_por_categoria = dados.groupby(['Loja', 'Categoria do Produto']).size().reset_index(name='Quantidade de Vendas')
vendas_populares = vendas_por_categoria.sort_values(['Loja', 'Quantidade de Vendas'], ascending=[True, False])
top3_categorias = vendas_populares.groupby('Loja').head(3)
top3_categorias
```

# 3. Média de Avaliação das Lojas

```python
# Calcular a média de avaliação por loja
media_avaliacao = dados.groupby('Loja')['Avaliação da compra'].mean().sort_values()

# Exibir a média das avaliações por loja
print(media_avaliacao)

```

# 5. Frete Médio por Loja
